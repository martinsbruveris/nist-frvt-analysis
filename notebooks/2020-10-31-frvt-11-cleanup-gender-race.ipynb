```python
import numpy as np
import pandas as pd
```

## Data cleanup

```python
df = pd.read_csv(
    "../data/frvt_11_gender_race_raw.csv", 
    header=None, 
    dtype=str
)
```

```python
df.head()
```

```python
len(df)
```

```python
# Some cells contain two numbers separated by space
# We split them into two cells
for j in range(len(df)):
    for k in range(len(df.columns)):
        if isinstance(df.iloc[j, k], str) and " " in df.iloc[j, k]:
            a, b = df.iloc[j, k].split()
            df.iloc[j, k] = a
            df.iloc[j, k+1] = b
```

```python
df.head()
```

```python
# Data latout is
#        k     k+1    k+2
#   j  NaN    Name    NaN
# j+1  NaN   MW_MB  MW_MW
# j+2  NaN   MB_MB  MW_MW
# j+3 FW_FB  FW_FW    NaN
# j+4 FB_FB  FB_FW    NaN
#
# Except when k=6, when we have
#        k     k+1
#   j  Name    NaN
# j+1 MW_MB  MW_MW
# j+2 MB_MB  MW_MW
# j+3 FW_FB  FW_FW
# j+4 FB_FB  FB_FW
data = {}
for j in range(0, len(df), 5):
    for k in [0, 3, 6, 8, 11]:
        if k == 6:
            name = df.iloc[j, k]
            entry = {
                "MW_MB": df.iloc[j+1, k],
                "MW_MW": df.iloc[j+1, k+1],
                "MB_MB": df.iloc[j+2, k],
                "MB_MW": df.iloc[j+2, k+1],
                "FW_FB": df.iloc[j+3, k],
                "FW_FW": df.iloc[j+3, k+1],
                "FB_FB": df.iloc[j+4, k],
                "FB_FW": df.iloc[j+4, k+1],
        }
        else:
            name = df.iloc[j, k+1]
            entry = {
                "MW_MB": df.iloc[j+1, k+1],
                "MW_MW": df.iloc[j+1, k+2],
                "MB_MB": df.iloc[j+2, k+1],
                "MB_MW": df.iloc[j+2, k+2],
                "FW_FB": df.iloc[j+3, k],
                "FW_FW": df.iloc[j+3, k+1],
                "FB_FB": df.iloc[j+4, k],
                "FB_FW": df.iloc[j+4, k+1],
            }
        name = name.replace("_", "-")
        entry = {key: float(value) for key, value in entry.items()}
        data[name] = entry
        
```

```python
res = pd.DataFrame(data).transpose()
```

```python
res.to_csv("../data/frvt_11_gender_race.csv")
```
